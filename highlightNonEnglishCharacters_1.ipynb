# Highlight Non English Characters v1

```python
file_path = r'C:\Users\srobi\OneDrive\Desktop\Temp\nonEnglishCharactersCheck.txt'
```

```python
import string
from IPython.display import display, HTML

# Define a function to highlight non-English characters in green
def highlight_non_english_characters(text):
    english_chars = set(string.ascii_letters + string.digits + string.punctuation + string.whitespace)
    
    # Create an empty list to store formatted text
    highlighted_text = []
    
    for char in text:
        if char not in english_chars:
            # Highlight non-English characters in green
            highlighted_text.append(f'<span style="background-color:green;">{char}</span>')
        else:
            highlighted_text.append(char)  # Keep English characters as is
    
    # Join the list into a single string
    highlighted_string = ''.join(highlighted_text)

    # Add HTML breaks
    highlighted_string = highlighted_string.replace('\n', '<br>')
    
    return highlighted_string

# # Read the contents of the text file
# file_path = "path_to_your_text_file.txt"

with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()

# Highlight non-English characters
highlighted_text = highlight_non_english_characters(text)

# Display the highlighted text in a Jupyter notebook
display(HTML(f"<p>{highlighted_text}</p>"))
```
